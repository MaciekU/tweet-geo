```python
import pandas as pd
import re
import numpy as np
import math
import os
from operator import itemgetter
import folium
from folium.plugins import HeatMap
```

```python
excel_data = pd.read_excel("UK-Geolocated-tweets-free-dataset-Followthehashtag/export_dashboard_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_10_32_03.xlsx", sheet_name="Stream")
```

```python

clean_data = excel_data[['Tweet content', 'Latitude', 'Longitude']]
```

```python
tags_geo = [] #tag + LAT LON
for index, row in clean_data.iterrows():
    if not math.isnan(row.Latitude) and not math.isnan(row.Longitude):
        tags = re.findall('#[a-zA-Z0-9]*', row['Tweet content'])
        tags_geo.extend([[tag, row.Latitude, row.Longitude] for tag in tags])#dodanie do tag√≥w lokalizacji
```

```python
tag = [x[0] for x in tags_geo] #x[0] tag
count_tag = [[x,tag.count(x)] for x in set(tag)]
```

```python
tag_by_popularity = sorted(count_tag, key=itemgetter(1), reverse=True)
```

```python
search_hash = "#nowplaying"
data = [[x[1], x[2]] for x in tags_geo if x[0] == search_hash]
map = folium.Map([54., -2.5], zoom_start=5)
HeatMap(data).add_to(map)
map
```
